### Step 3. Assign it to a variable called data and replaced the first 3 columns by a proper datetime index.

```python
data = pd.read_csv("wind.txt", sep = "\s+", parse_dates = {'Date':[0,1,2]})
data.head()
```

### Step 4. Do we really have data from year 2061

```python
from datetime import date

# func
def fix_date(x):
    if x.year >1989:
        year = x.year - 100
    else:
        year = x.year
    return date(year, x.month, x.day)
      
data["Date"] = data["Date"].apply(fix_date)

data.head()
```

### Step 5. The right dates as the index

```python
data.info()
```

```python
# data type : datetime64[ns]
data["Date"] = data["Date"].astype('datetime64[ns]')

data.info()
```

```python
# dates as the index

data = data.set_index('Date')
data.head()
```

### Step 6. Values which are missing for each location over the entire record. 

```python
data.isnull().sum()
```

### Step 7. Non-missing values

```python
data.shape[0] - data.isnull().sum()
```

### Step 8. The mean windspeeds of the windspeeds over all the locations and all the times.

```python
mean_wind = data.sum().sum() / data.notna().sum().sum()

mean_wind
```

### Step 9. A dataFrame called loc_stats and calculated the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

```python
def new_stats(x):
    x_ser = pd.Series(x)
    x_Min = x_ser.min()
    x_Max = x_ser.max()
    x_Mean = x_ser.mean()
    x_Std = x_ser.std()
    result = [x_Min,x_Max,x_Mean,x_Std]
    ind = ["Min","Max","Mean","Std"]
    result = pd.Series(result,index=ind)
    return result

loc_stats = data.apply(new_stats)
loc_stats
```

### Step 10. A dataFrame called day_stats and calculated the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day

```python
day_stats = data.apply(new_stats,axis=1)
day_stats.head()
```

### Step 11. Average windspeed in January for each location by treating January 1961 and January 1962 both as January

```python
january_wind = data[data.index.month == 1]

print("January windspeeds:")
january_wind.mean()
```

### Step 12. The record to a yearly frequency for each location.

```python
data.resample('A').mean()
```

### Step 13. The record to a monthly frequency for each location.

```python
data.resample('M').mean()
```

### Step 14. The record to a weekly frequency for each location.

```python
data.resample('W').mean()
```

### Step 15. The min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week for first 52 weeks

```python
each_week = data.groupby(data.index.to_period("W")).agg(["min", "max", "mean", "std"])
each_week.iloc[1:53, :].head()
```

```python
import pandas as pd

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data = data[data['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = data['Year'].values
marriages = data['Marriages_per_1000'].values
divorces = data['Divorces_per_1000'].values
```

```python
import matplotlib.pyplot as plt

years = years
marriages = marriages
divorces = divorces

plt.figure()

plt.bar(years - 10,
        marriages,
        color='blue',
        edgecolor='none',
        width=20,
        align='center',
        label='Marriages')

plt.bar(years + 10,
        divorces,
        color='orange',
        edgecolor='none',
        width=20,
        align='center',
        label='Divorces')

plt.xticks(years, [str(year) for year in years])

plt.xlabel("Years")
plt.ylabel("Marriages/Divorces per 1000")
plt.legend()
```

```python
import pandas as pd

roman_emp = pd.read_csv('roman-emperor-reigns.csv')
assassinated_emp = roman_emp[
    roman_emp['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]

n_assassinated = len(assassinated_emp)
othr_death = len(roman_emp) - n_assassinated
```

```python
import matplotlib.pyplot as plt

counts = [n_assassinated, othr_death]

plt.figure(figsize=(4, 4))

plt.pie(counts,
        colors=['blue', 'orange'],
        labels=['Assassinated', 'Other deaths'],
        startangle=90,
        autopct='%1.1f%%')
```
